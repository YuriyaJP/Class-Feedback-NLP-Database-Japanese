```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline

# data with labels
data = [
    {"text": "いろんな 国 いい ところ 知れ た 国 調べる 楽しかっ た", "label": "Positive"},
    {"text": "他 国 教育 システム 理解 でき た", "label": "Positive"},
    {"text": "世界 各国 多様 性 尊重 する こと 大切 だ と 感じ た", "label": "Positive"},
    {"text": "新た な 友人 でき 嬉しかっ た", "label": "Positive"},
    {"text": "異国 の 食事 が 美味しかっ た", "label": "Positive"},
    {"text": "素晴らしい 文化 体験 が でき た", "label": "Positive"},
    {"text": "国際交流 は 楽しい 体験 だっ た", "label": "Positive"},
    # negative examples
    {"text": "言葉 の 違い が 障害 と なっ た", "label": "Negative"},
    {"text": "異文化 コミュニケーション が 難しかっ た", "label": "Negative"},
    {"text": "文化 衝突 が 起こっ た", "label": "Negative"},
]

# text and labels
texts = [item["text"] for item in data]
labels = [item["label"] for item in data]

# a LabelEncoder to convert text labels to numeric values
label_encoder = LabelEncoder()
numeric_labels = label_encoder.fit_transform(labels)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(texts, numeric_labels, test_size=0.2, random_state=42)

# Create a TF-IDF vectorizer and a Multinomial Naive Bayes classifier
tfidf_vectorizer = TfidfVectorizer()
classifier = MultinomialNB()

# Create a classification pipeline
text_classifier = Pipeline([
    ('tfidf', tfidf_vectorizer),
    ('classifier', classifier),
])

# Train the classifier on the training data
text_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = text_classifier.predict(X_test)

# Convert the numeric predictions back to text labels
predicted_labels = label_encoder.inverse_transform(y_pred)

# Calculate accuracy and print a classification report
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{classification_rep}")
```
